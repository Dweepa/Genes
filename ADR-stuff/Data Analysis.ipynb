```python
classifierNames = ["NB","DT", "RF", "SGD" , "LR"]
lims = [10, 20, 30, 50, 100, 200, 300, 500, 800, 1000, 1300, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000]
```

```python
afib = pd.DataFrame(columns=['Model', 'ADR', 'PCA', 'Avg F1', 'Avg Recall', 'Avg Precision', 'Avg ROC'])
for lim in lims:
    for clf in classifierNames:
        df1 = pd.read_csv("College/Results/AFIB/PCA-" + str(lim) + "/" + clf + "/Averages/" + clf + "_AFIB_Average.csv")
        df1['Model'] = clf
        afib = pd.concat([afib, df1], axis=0, ignore_index=True)
afib.drop(["Unnamed: 0", "Avg Acc"], axis=1, inplace=True)
afib
```

```python
df = pd.DataFrame()
for clf in classifierNames:   
    n = afib.loc[afib['Model'] == clf]
    n["Avg F1"] = n["Avg F1"]/5
    n.loc[n['PCA'] == 1000, "Avg F1"] *= 5
    n.sort_values(by=['PCA'], inplace=True)
    df = pd.concat([df, n], axis=0, ignore_index=True)
```
